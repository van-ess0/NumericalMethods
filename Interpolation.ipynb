```python
import numpy as np
from math import asin, pi
import matplotlib.pyplot as plt
from pylab import *
```

```python
def function1(x):
    return x ** 2 - 180 * asin(x) / pi


vfunc1 = np.vectorize(function1)
xnew = np.linspace(-1, 1, 300)
yold = vfunc1(xnew)
```

```python
X1 = np.linspace(-1, 1, 11)
Y1 = vfunc1(X1)
print(list(zip(X1, Y1)))
```

```python
def lagrange(points):
    def P(x):
        total = 0
        n = len(points)
        for i in range(n):
            xi, yi = points[i]

            def g(i, n):
                tot_mul = 1
                for j in range(n):
                    if i == j:
                        continue
                    xj, yj = points[j]
                    tot_mul *= (x - xj) / (xi - xj)

                return tot_mul

            total += yi * g(i, n)
        return total
    return P
```

```python
P1 = lagrange(list(zip(X1, Y1)))
ynew = np.array(list(map(P1, xnew)))
plt.plot(X1, Y1, 'o', xnew, ynew, xnew, yold, '--')
plt.grid(True)
plt.show()
```

```python
X2 = np.linspace(-1, 1, 33)
Y2 = vfunc1(X2)
```

```python
P2 = lagrange(list(zip(X2, Y2)))
ynew2 = np.array(list(map(P2, xnew)))
plt.plot(X2, Y2, 'o', xnew, ynew2, xnew, yold, '--')
plt.grid(True)
plt.show()
```
